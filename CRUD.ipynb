```python
lst = []

while True:
    print("\n1. Create")
    print("2. Read")
    print("3. Update")
    print("4. Delete")
    print("5. Exit")

    choice = input("Enter choice: ").strip()

    if choice == "1":
        lst = input("Enter elements : ").split()

    elif choice == "2":
        print(lst)

    elif choice == "3":
        old = input("Enter element to update: ").strip()
        if old in lst:
            new = input("Enter new value: ").strip()
            lst[lst.index(old)] = new
        else:
            print("Not found:", old)

    elif choice == "4":
        item = input("Enter element to delete: ").strip()
        if item in lst:
            lst.remove(item)
        else:
            print("Not found:", item)

    elif choice == "5":
        break
```

```python
lst = input("Enter 5 elements: ").split()

while True:
    print("\n1.Create \n2.Read \n3.Update \n4.Delete \n5.Exit")
    c = input()

    if c == "1":
        lst += input().split()

    elif c == "2":
        print(lst)

    elif c == "3":
        i = input()
        if i in lst:
            lst[lst.index(i)] = input()

    elif c == "4":
        i = input()
        if i in lst:
            lst.remove(i)

    elif c == "5":
        break
```

```python
lst = ["1", "2", "3", "4", "5"]

while True:
    print("\n1.Create(Add at end) \n2.Read \n3.Update \n4.Delete \n5.Exit")
    c = input("Enter choice: ")

    if c == "1":
        lst.append(input("Enter element to add at end: "))

    elif c == "2":
        print("List:", lst)

    elif c == "3":
        i = input("Enter element to update: ")
        if i in lst:
            lst[lst.index(i)] = input("Enter new value: ")

    elif c == "4":
        i = input("Enter element to delete: ")
        if i in lst:
            lst.remove(i)

    elif c == "5":
        break

print("Final List:", lst)
```

```python
d = {"1": "A", "2": "B", "3": "C"}

while True:
    print("\n1.Create 2.Read 3.Update 4.Delete 5.Exit")
    c = input("Enter choice: ")

    if c == "1":
        k = input("Enter key: ")
        v = input("Enter value: ")
        d[k] = v

    elif c == "2":
        print(d)

    elif c == "3":
        k = input("Enter key to update: ")
        if k in d:
            d[k] = input("Enter new value: ")

    elif c == "4":
        k = input("Enter key to delete: ")
        if k in d:
            d.pop(k)

    elif c == "5":
        break

print("Final Dictionary:", d)
```
